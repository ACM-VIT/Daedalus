```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_samples = []
train_labels = []
```

```python
for i in range(1000):
    x = randint(13,64)
    train_samples.append(x)
    train_labels.append(0)
    
    y = randint(65,100)
    train_samples.append(y)
    train_labels.append(1)
    
for i in range(50):
    x = randint(13,64)
    train_samples.append(x)
    train_labels.append(1)
    
    y = randint(65,100)
    train_samples.append(y)
    train_labels.append(0)
```

```python
train_samples = np.array(train_samples)
train_label = np.array(train_labels)
print(train_samples.shape)
for i in train_samples:
    print(i)
```

```python
scaler = MinMaxScaler(feature_range = (0,1))
scaled_train_sample = scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
for i in scaled_train_sample:
    print(i)
```

```python
import os

with open(os.path.expanduser('~')+'\\.keras\\keras.json','w') as f:
    new_settings = """{\r\n
    "epsilon": 1e-07,\r\n
    "image_data_format": "channels_last",\n
    "backend": "tensorflow",\r\n
    "floatx": "float32"\r\n
    }"""
    f.write(new_settings)

import keras
from keras import backend as t
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([Dense(16,input_shape=(1,),activation = 'relu'),
                   Dense(32,activation='relu'),
                   Dense(2,activation='sigmoid')
                   ])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(lr=0.01),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(scaled_train_sample,train_labels,validation_split=0.4,batch_size=10,epochs=20,verbose=2)
```
